4. Next we want to see if there's any correlation between the Growth Rate and Total Fertility Rate.<br>

4a. __Use an appropriate plot to show whether there is any correlation__.

Make sure your plot has a title, x-axis and y-axis labels.

5. Next __find the 12 countries with the highest Population Density__<br>
and __use an appropriate plot to show the Population Density and Names__, sorted by Population Density.

The plot should have a title, country Names, and corresponding Population Density.<br>
The country Names should not overlap each other so that they're readable.

```python
top_countries = world_population.sort_values(by='Population Density', ascending=False).head(12)

plt.bar(top_countries['Name'],top_countries['Population Density'], color='skyblue')
plt.xticks(rotation=45, fontsize=8)
plt.yticks(fontsize=8)
plt.xlabel("Country Names", weight='bold')
plt.ylabel("Population Density", weight='bold')
plt.title("Top 12 Countries with the Highest Population Density")
```

7. Last we look at the ratio of Working Age people vs Youth and Old Age.<br>

7a. __Print the country Names and 'Old Age and Working Age Ratio'__ of countries that are within the top 5% of Old Age and Working Age Ratio.<br>

_Note 1: the question asked for the top 5% (not the top 5) of the ratio.<br>

```python
top_5_percent = np.percentile(world_population['Old Age and Working Age Ratio'], 95)
result = world_population[world_population['Old Age and Working Age Ratio'] >= top_5_percent][['Name', 'Old Age and Working Age Ratio']]
result_sorted = result.sort_values(by='Old Age and Working Age Ratio', ascending=False)
result_sorted
```
